```python
mapping = {
    'COMPACT': 'Four-Wheeler',
    'SUV - SMALL': 'Four-Wheeler',
    'MID-SIZE': 'Four-Wheeler',
    'TWO-SEATER': 'Four-Wheeler',
    'MINICOMPACT': 'Four-Wheeler',
    'SUBCOMPACT': 'Four-Wheeler',
    'FULL-SIZE': 'Four-Wheeler',
    'STATION WAGON - SMALL': 'Four-Wheeler',
    'SUV - STANDARD': 'Four-Wheeler',
    'STATION WAGON - MID-SIZE': 'Four-Wheeler',
    'VAN - CARGO': 'Commercial',
    'VAN - PASSENGER': 'Commercial',
    'PICKUP TRUCK - STANDARD': 'Commercial',
    'PICKUP TRUCK - SMALL': 'Commercial',
    'MINIVAN': 'Three-Wheeler',
    'SPECIAL PURPOSE VEHICLE': 'Commercial'
}
df['Vehicle Class'] = df['Vehicle Class'].map(mapping)
```

```python
df
```

```python
from sklearn.preprocessing import MinMaxScaler
features = ['Fuel Consumption Hwy (L/100 km)', 
            'Fuel Consumption Comb (L/100 km)',
            'Fuel Consumption Comb (mpg)',
            'CO2 Emissions(g/km)']

scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(df[features])
vehicle_age_score = scaled_features.mean(axis=1)
df['vehicle_age'] = np.ceil(vehicle_age_score * 9 + 1).astype(int)
```

```python
df['Fuel Type'].value_counts()
```

```python
fuel_mapping = {
    'X': 'Petrol',
    'Z': 'Diesel',
    'E': 'Electric',
    'D': 'CNG',
    'N': 'Hybrid'
}

df['Fuel Type'] = df['Fuel Type'].map(fuel_mapping)
```

```python
df.rename(columns={"Engine Size(L)": "Engine_Capacity"}, inplace = True)
df.rename(columns={"Vehicle Class": 'Vehicle_Type'}, inplace = True)
```

```python
def estimate_distance(row):
    base_distance = 500    
    base_distance += row['Engine_Capacity'] * 100      
    if row['Fuel Type'] in ['Diesel']:
        base_distance += 200    
    if row['vehicle_age'] > 5:
        base_distance -= 100    
    if row['Fuel Type'] in ['Electric', 'Hybrid']:
        base_distance -= 100
    return base_distance
df['Distance_Driven_Last_Month'] = df.apply(estimate_distance, axis=1)
print(df[['Make', 'Model', 'Distance_Driven_Last_Month']])
    
```

```python
df['Fuel_Consumed_Last_Month'] = (df['Distance_Driven_Last_Month'] * 
                                  df['Fuel Consumption Comb (L/100 km)']) / 100
```
